## Enunciado

Pretende-se um programa que leia um dataset de obras, processe-o e crie os seguintes resultados:
* Lista ordenada alfabeticamente dos compositores musicais;
* Distribuição das obras por período: quantas obras catalogadas em cada período;
* Dicionário em que a cada período está a associada uma lista alfabética dos títulos das obras desse período.

## Resolução

```python
import re

def getCompositores(path):
    with open(path) as file:
        file.readline()

        pattern = r'\s[^;]*?(?:;\"\"\"|;\")[^;]+(?:\"\"\";|\";)[^;]*?;[^;]*?;([^;]*?);[^;]*?;.*'

        l = set(re.findall(pattern,file.read(),re.MULTILINE))
        l = sorted(l)

        return l
```

```python
import re

def obrasPorPeriodo(path):
    d = {}
    l = []
    with open(path) as file:
        file.readline()

        pattern = r'\s[^;]*?(?:;\"?(?:\"\"|[^\"])*?\"?;)\d{4};([^;]*?);[^;]*?;[^;]*?;\d*?'

        l = re.findall(pattern,file.read(),re.MULTILINE)
        
        for x in l:
            if x not in d:
                d[x] = l.count(x)

    return d
```

```python
import re

def DictObrasPorPeriodo(path):
    d = {}
    l = []
    with open(path) as file:
        file.readline()

        pattern = r'\s([^;]*?)(?:;\"?(?:\"\"|[^\"])*?\"?;)\d{4};([^;]*?);[^;]*?;[^;]*?;\d*?'

        l = re.findall(pattern,file.read(),re.MULTILINE)
        
        for x in l:
            if x[1] in d: d[x[1]].append(x[0])
            else: d[x[1]] = [x[0]]

        for y in d:
            d[y].sort()

    return d
```

```python
def obrasStats(path):
        return(getCompositores(path),obrasPorPeriodo(path),DictObrasPorPeriodo(path))
```
